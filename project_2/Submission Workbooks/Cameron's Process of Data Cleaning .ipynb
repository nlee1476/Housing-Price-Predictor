```python
def run_model(data, features, target = 'SalePrice', 
              model=LinearRegression):
    """
    Run model function 
    
    Inputs:
        data - pandas dataframe
        model - model algorithm of choice
        features (X variables) - list of features/columns in data
        target (y variable) - target variable column (string)
    """
    X = data[features]
    y = data[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True,
                                                        random_state = 42)
    
    # this might not work
    model = model()
    
    model.fit(X_train, y_train)
    print(f'Train score: {model.score(X_train, y_train)}')
    print(f'Cross-val score: {cross_val_score(model, X_train, y_train, cv=5).mean()}')
    print(f'Test score: {model.score(X_test, y_test)}')
    
    return model
```
